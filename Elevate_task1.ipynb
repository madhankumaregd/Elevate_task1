```python
# Statistical summary of numerical columns
print("STATISTICAL SUMMARY:")
print("="*50)
df.describe()
```

```python
# Check missing values in each column
print("MISSING VALUES COUNT:")
print("="*50)
missing_values = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100

# Create a dataframe to display missing values nicely
missing_df = pd.DataFrame({
    'Column': df.columns,
    'Missing Count': missing_values,
    'Missing %': missing_percentage
})
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing Count', ascending=False)

if len(missing_df) > 0:
    print(missing_df)
else:
    print("✅ No missing values found!")
    
print(f"\nTotal missing values: {df.isnull().sum().sum()}")
```

```python
# Check for duplicate rows
print("DUPLICATE ROWS CHECK:")
print("="*50)
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

if duplicate_count > 0:
    # Show first few duplicate rows
    print("\nSample of duplicate rows:")
    df[df.duplicated(keep=False)].sort_values(by=list(df.columns)).head(10)
else:
    print("✅ No duplicate rows found!")
```

```python
# Check data types of each column
print("DATA TYPES:")
print("="*50)
dtype_df = pd.DataFrame({
    'Column': df.columns,
    'Current Data Type': df.dtypes.values
})
print(dtype_df)
```

```python
# Check unique values in each column
print("UNIQUE VALUES COUNT:")
print("="*50)
for col in df.columns:
    unique_count = df[col].nunique()
    print(f"{col}: {unique_count} unique values")
    
    # For categorical columns with few unique values, show the values
    if unique_count < 20:
        print(f"   Values: {sorted(df[col].unique())}")
    print()
```

```python
# Create a copy to work on
df_clean = df.copy()
print("✅ Working copy created!")
print(f"Original shape: {df.shape}")
print(f"Copy shape: {df_clean.shape}")
```

```python
# Clean column names - lowercase, no spaces, replace special characters
print("ORIGINAL COLUMN NAMES:")
print(df_clean.columns.tolist())

# Clean column names
df_clean.columns = df_clean.columns.str.lower()
df_clean.columns = df_clean.columns.str.replace('-', '_')
df_clean.columns = df_clean.columns.str.replace(' ', '_')
df_clean.columns = df_clean.columns.str.replace('/', '_')

print("\n✅ CLEANED COLUMN NAMES:")
print(df_clean.columns.tolist())
```

```python
# Convert ID columns to strings (they are identifiers, not numbers)
print("BEFORE - Data types:")
print(f"patientid: {df_clean['patientid'].dtype}")
print(f"appointmentid: {df_clean['appointmentid'].dtype}")

# Convert to string
df_clean['patientid'] = df_clean['patientid'].astype(str)
df_clean['appointmentid'] = df_clean['appointmentid'].astype(str)

print("\n✅ AFTER - Data types:")
print(f"patientid: {df_clean['patientid'].dtype}")
print(f"appointmentid: {df_clean['appointmentid'].dtype}")
```

```python
# Convert date columns to datetime
print("DATE COLUMNS CONVERSION:")
print("="*50)

# Check original format
print("Sample of original dates:")
print(df_clean[['scheduledday', 'appointmentday']].head(3))

# Convert to datetime
df_clean['scheduledday'] = pd.to_datetime(df_clean['scheduledday'], format='%Y-%m-%dT%H:%M:%SZ', errors='coerce')
df_clean['appointmentday'] = pd.to_datetime(df_clean['appointmentday'], format='%Y-%m-%dT%H:%M:%SZ', errors='coerce')

print("\n✅ AFTER conversion:")
print(df_clean[['scheduledday', 'appointmentday']].head(3))
print(f"\nData types after conversion:")
print(df_clean[['scheduledday', 'appointmentday']].dtypes)
```

```python
# Extract additional date components
print("EXTRACTING DATE COMPONENTS:")
print("="*50)

# Extract date (without time)
df_clean['scheduled_date'] = df_clean['scheduledday'].dt.date
df_clean['appointment_date'] = df_clean['appointmentday'].dt.date

# Extract time
df_clean['scheduled_time'] = df_clean['scheduledday'].dt.time

# Extract day of week
df_clean['scheduled_dayofweek'] = df_clean['scheduledday'].dt.day_name()
df_clean['appointment_dayofweek'] = df_clean['appointmentday'].dt.day_name()

# Extract month
df_clean['scheduled_month'] = df_clean['scheduledday'].dt.month
df_clean['appointment_month'] = df_clean['appointmentday'].dt.month

print("✅ New columns created:")
print(df_clean[['scheduled_date', 'appointment_date', 'scheduled_time', 
                'scheduled_dayofweek', 'appointment_dayofweek']].head())
```

```python
# Calculate wait time between scheduling and appointment
print("CALCULATING WAIT TIME:")
print("="*50)

df_clean['wait_days'] = (df_clean['appointmentday'] - df_clean['scheduledday']).dt.days

print("Wait time statistics (in days):")
print(df_clean['wait_days'].describe())

# Check for negative wait times (data errors)
negative_wait = df_clean[df_clean['wait_days'] < 0]
print(f"\nRows with negative wait time: {len(negative_wait)}")

if len(negative_wait) > 0:
    print("\nSample of negative wait times:")
    print(negative_wait[['patientid', 'scheduledday', 'appointmentday', 'wait_days']].head())
    
    # Fix negative wait times by taking absolute value
    df_clean['wait_days'] = df_clean['wait_days'].abs()
    print("\n✅ Fixed negative wait times by taking absolute value")
```

```python
# Standardize gender values
print("GENDER COLUMN BEFORE:")
print(df_clean['gender'].value_counts())

# Standardize to uppercase
df_clean['gender'] = df_clean['gender'].str.upper().str.strip()

# Check for invalid values
valid_genders = ['F', 'M']
invalid_genders = df_clean[~df_clean['gender'].isin(valid_genders)]

print(f"\nInvalid gender values: {len(invalid_genders)}")
if len(invalid_genders) > 0:
    print(invalid_genders['gender'].value_counts())
    # Fill invalid with most common gender
    most_common = df_clean['gender'].mode()[0]
    df_clean.loc[~df_clean['gender'].isin(valid_genders), 'gender'] = most_common
    print(f"✅ Filled invalid genders with '{most_common}'")

print("\nGENDER COLUMN AFTER:")
print(df_clean['gender'].value_counts())
```

```python
# Check age column for invalid values
print("AGE COLUMN ANALYSIS:")
print("="*50)

print("Age statistics:")
print(df_clean['age'].describe())

# Check for negative ages
negative_age = df_clean[df_clean['age'] < 0]
print(f"\nNegative ages: {len(negative_age)}")

# Check for unrealistic ages (> 100)
old_age = df_clean[df_clean['age'] > 100]
print(f"Ages > 100: {len(old_age)}")

# Check for ages > 120 (extremely unrealistic)
very_old = df_clean[df_clean['age'] > 120]
print(f"Ages > 120: {len(very_old)}")

# Handle invalid ages
if len(negative_age) > 0 or len(very_old) > 0:
    median_age = df_clean[(df_clean['age'] >= 0) & (df_clean['age'] <= 100)]['age'].median()
    print(f"\nMedian age (for valid data): {median_age}")
    
    # Replace invalid ages with median
    df_clean.loc[(df_clean['age'] < 0) | (df_clean['age'] > 120), 'age'] = median_age
    print(f"✅ Replaced {len(negative_age) + len(very_old)} invalid ages with median")

print("\nAge statistics after cleaning:")
print(df_clean['age'].describe())
```

```python
# Create age groups for analysis
print("CREATING AGE GROUPS:")
print("="*50)

# Define age bins and labels
bins = [0, 18, 35, 50, 65, 100, 120]
labels = ['0-17', '18-34', '35-49', '50-64', '65-99', '100+']

df_clean['age_group'] = pd.cut(df_clean['age'], bins=bins, labels=labels, right=False)

print("Age group distribution:")
print(df_clean['age_group'].value_counts().sort_index())
```

```python
# Standardize neighbourhood names
print("NEIGHBOURHOOD COLUMN:")
print("="*50)

print(f"Unique neighbourhoods before: {df_clean['neighbourhood'].nunique()}")

# Clean neighbourhood names
df_clean['neighbourhood'] = df_clean['neighbourhood'].str.strip().str.title()

# Replace common variations (if any)
neighbourhood_mapping = {
    # Add any corrections here if needed
    # 'Jardim Da Penha': 'Jardim da Penha',
}

if neighbourhood_mapping:
    df_clean['neighbourhood'] = df_clean['neighbourhood'].replace(neighbourhood_mapping)

print(f"Unique neighbourhoods after: {df_clean['neighbourhood'].nunique()}")
print("\nTop 10 neighbourhoods:")
print(df_clean['neighbourhood'].value_counts().head(10))
```

```python
# Clean binary columns (should be 0 or 1)
binary_cols = ['scholarship', 'hipertension', 'diabetes', 'alcoholism', 'handcap', 'sms_received']

print("BINARY COLUMNS CHECK:")
print("="*50)

for col in binary_cols:
    print(f"\n{col.upper()} - Unique values before:")
    print(df_clean[col].unique())
    
    # Ensure values are 0 or 1 (any positive value becomes 1)
    df_clean[col] = df_clean[col].apply(lambda x: 1 if x > 0 else 0 if pd.notna(x) else 0)
    
    print(f"After cleaning - Unique values:")
    print(df_clean[col].unique())
    print(f"Value counts: 0: {(df_clean[col] == 0).sum()}, 1: {(df_clean[col] == 1).sum()}")
```

```python
# Handicap column can have values 0-4, check distribution
print("HANDICAP COLUMN ANALYSIS:")
print("="*50)

print("Value counts:")
print(df_clean['handcap'].value_counts().sort_index())

# Check for values > 4
invalid_handcap = df_clean[df_clean['handcap'] > 4]
print(f"\nRows with handcap > 4: {len(invalid_handcap)}")

if len(invalid_handcap) > 0:
    print("Values found:", invalid_handcap['handcap'].unique())
    # Cap at 4
    df_clean['handcap'] = df_clean['handcap'].apply(lambda x: 4 if x > 4 else x)
    print("✅ Capped handcap values at 4")
```

```python
# Standardize no-show column
print("NO-SHOW COLUMN BEFORE:")
print(df_clean['no_show'].value_counts())

# Standardize to Yes/No
df_clean['no_show'] = df_clean['no_show'].str.strip().str.capitalize()

# Create binary column for analysis
df_clean['no_show_binary'] = df_clean['no_show'].map({'No': 0, 'Yes': 1})

print("\nNO-SHOW COLUMN AFTER:")
print(df_clean['no_show'].value_counts())
print(f"\nNo-show rate: {df_clean['no_show_binary'].mean():.2%}")
```

```python
# Remove duplicate rows
print("DUPLICATE ROWS:")
print("="*50)

initial_rows = len(df_clean)
print(f"Rows before removing duplicates: {initial_rows}")

# Check for duplicates based on key columns
duplicates = df_clean.duplicated()
duplicate_count = duplicates.sum()
print(f"Duplicate rows found: {duplicate_count}")

if duplicate_count > 0:
    # Remove duplicates (keep first occurrence)
    df_clean = df_clean.drop_duplicates(keep='first')
    print(f"Rows after removing duplicates: {len(df_clean)}")
    print(f"✅ Removed {initial_rows - len(df_clean)} duplicate rows")
else:
    print("✅ No duplicates to remove")
```

```python
# Check for patients with same info but different appointment IDs
print("CHECKING FOR NEAR-DUPLICATES:")
print("="*50)

# Group by all columns except appointmentid and patientid
non_id_cols = ['gender', 'age', 'neighbourhood', 'scholarship', 'hipertension', 
               'diabetes', 'alcoholism', 'handcap', 'sms_received', 'no_show']

# Find groups that appear more than once
duplicate_groups = df_clean.groupby(non_id_cols).size().reset_index(name='count')
duplicate_groups = duplicate_groups[duplicate_groups['count'] > 1]

print(f"Found {len(duplicate_groups)} groups of identical patient information")
print("These are likely different appointments for the same patient type")
```

```python
# Check for missing values again after cleaning
print("MISSING VALUES AFTER PREVIOUS STEPS:")
print("="*50)
missing_after = df_clean.isnull().sum()
print(missing_after[missing_after > 0])

# Handle remaining missing values
print("\nHANDLING REMAINING MISSING VALUES:")
print("="*50)

# 1. Drop rows with missing dates (critical info)
date_cols = ['scheduledday', 'appointmentday']
before_drop = len(df_clean)
df_clean = df_clean.dropna(subset=date_cols)
print(f"Dropped {before_drop - len(df_clean)} rows with missing dates")

# 2. Fill missing neighbourhood with 'Unknown'
if df_clean['neighbourhood'].isnull().sum() > 0:
    df_clean['neighbourhood'] = df_clean['neighbourhood'].fillna('Unknown')
    print(f"✅ Filled {df_clean['neighbourhood'].isnull().sum()} missing neighbourhoods")

# 3. Fill missing gender with mode
if df_clean['gender'].isnull().sum() > 0:
    mode_gender = df_clean['gender'].mode()[0]
    df_clean['gender'] = df_clean['gender'].fillna(mode_gender)
    print(f"✅ Filled missing genders with '{mode_gender}'")

# 4. Fill missing binary columns with 0
for col in binary_cols:
    if col in df_clean.columns and df_clean[col].isnull().sum() > 0:
        df_clean[col] = df_clean[col].fillna(0)
        print(f"✅ Filled missing {col} with 0")

# Final check
print("\nFINAL MISSING VALUES CHECK:")
print(df_clean.isnull().sum())
```

```python
# Check and fix data types
print("FINAL DATA TYPES:")
print("="*50)

# Define desired data types
df_clean['patientid'] = df_clean['patientid'].astype(str)
df_clean['appointmentid'] = df_clean['appointmentid'].astype(str)
df_clean['gender'] = df_clean['gender'].astype('category')
df_clean['neighbourhood'] = df_clean['neighbourhood'].astype('category')
df_clean['age'] = df_clean['age'].astype(int)
df_clean['no_show'] = df_clean['no_show'].astype('category')

# Binary columns to int
for col in binary_cols:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].astype(int)

print(df_clean.dtypes)
```

```python
# Create summary statistics for cleaned data
print("SUMMARY STATISTICS FOR CLEANED DATA:")
print("="*50)

# Overall stats
print(f"Total rows: {len(df_clean):,}")
print(f"Total columns: {len(df_clean.columns)}")
print(f"Date range: {df_clean['scheduledday'].min()} to {df_clean['scheduledday'].max()}")

# Patient stats
print(f"\nUnique patients: {df_clean['patientid'].nunique():,}")
print(f"Unique appointments: {df_clean['appointmentid'].nunique():,}")

# Gender distribution
print(f"\nGender distribution:")
print(df_clean['gender'].value_counts(normalize=True).mul(100).round(1))

# Age stats
print(f"\nAge - Mean: {df_clean['age'].mean():.1f}, Median: {df_clean['age'].median():.1f}")

# No-show rate
print(f"\nNo-show rate: {df_clean['no_show_binary'].mean():.2%}")

# Average wait time
print(f"Average wait time: {df_clean['wait_days'].mean():.1f} days")
```

```python
# Create some basic visualizations
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Age distribution
df_clean['age'].hist(ax=axes[0,0], bins=30, edgecolor='black')
axes[0,0].set_title('Age Distribution')
axes[0,0].set_xlabel('Age')
axes[0,0].set_ylabel('Count')

# Gender distribution
df_clean['gender'].value_counts().plot(kind='bar', ax=axes[0,1], color=['pink', 'blue'])
axes[0,1].set_title('Gender Distribution')
axes[0,1].set_xlabel('Gender')
axes[0,1].set_ylabel('Count')

# No-show by gender
pd.crosstab(df_clean['gender'], df_clean['no_show']).plot(kind='bar', ax=axes[0,2])
axes[0,2].set_title('No-Show by Gender')
axes[0,2].set_xlabel('Gender')
axes[0,2].set_ylabel('Count')

# Age groups
df_clean['age_group'].value_counts().sort_index().plot(kind='bar', ax=axes[1,0])
axes[1,0].set_title('Age Groups')
axes[1,0].set_xlabel('Age Group')
axes[1,0].set_ylabel('Count')

# Wait time distribution
df_clean['wait_days'].hist(ax=axes[1,1], bins=50, edgecolor='black')
axes[1,1].set_title('Wait Time Distribution')
axes[1,1].set_xlabel('Days')
axes[1,1].set_ylabel('Count')

# No-show by age
df_clean.groupby('age_group')['no_show_binary'].mean().plot(kind='bar', ax=axes[1,2])
axes[1,2].set_title('No-Show Rate by Age Group')
axes[1,2].set_xlabel('Age Group')
axes[1,2].set_ylabel('No-Show Rate')

plt.tight_layout()
plt.show()
```
